```python
Unique_Dates = [str(pd.to_datetime(i).date()) for i in list(data['Date'].unique())]
```

```python
%%time
import os
path = r"C:\Users\arushi.roy\Desktop\work1\BankNifty_MarketData\BankNifty_Data_Datewise"
error_dt = list()
prob_dt =list()
for dt in Unique_Dates:
    try:
        
        
        ThatDayDf = pd.read_csv(os.path.join(path,dt+'.csv'))
        ThatDayDf['DateTime'] = pd.to_datetime(ThatDayDf['DateTime']) 
        ThatDayDf['Time'] = ThatDayDf['DateTime'].dt.time

        Tmpdf = data[data['Date'] == pd.to_datetime(dt)].reset_index(drop=True)
        for i in range(len(Tmpdf)):
            ticker = Tmpdf['Ticker'][i]
            entry_time = Tmpdf['Entry Time'][i]
            exit_time = Tmpdf['Exit Time'][i]
            entry_point = Tmpdf['Entry Point'][i]
            exit_value = Tmpdf['Exit Value'][i]

            tocompare_entry_point  = ThatDayDf[(ThatDayDf['Ticker'] == ticker) & \
                                               (ThatDayDf['Time'] == pd.to_datetime(entry_time).time())]['Close'].values[0]
            tocompare_exit_value = ThatDayDf[(ThatDayDf['Ticker'] == ticker) & \
                                             (ThatDayDf['Time'] == pd.to_datetime(exit_time).time())]['Close'].values[0]
            res1=np.where(tocompare_entry_point ==entry_point,1,0)
            res2=np.where(tocompare_exit_value ==exit_value,1,0)
            if (res1 ==0 | res2 ==0):
                prob_dt.append(dt)
            print(dt)
    except Exception as e:
        print(dt,e)
        error_dt.append(dt)
        continue
```
